```python
# import the necessary libraries
import importlib
import turtle
from kolam_shapes import *
import numpy as np
from turtle import Screen
import tkinter
from tkinter import filedialog

importlib.reload(turtle)

# turtle details

pen = turtle.Turtle()
pen.shape("circle") # shape of turtle
pen.pensize(5) # size of turtle
pen.speed(2)   # speed of drawing

# pen at required co-ordinates

pen.penup()
pen.goto(-150, 200)
pen.pendown()

# Ratios for Kolam shapes 
# space/size(or distance of shape-II, III & IV) = 85/30 = 2.82, 
# space/(distance of shape-I) = 85/60 = 1.42, 



# method to draw square with dots
# space --> distance between dots
# x     --> side of square
def draw(x, space = 85):
  for i in range(x):
    for j in range(x):

        # dot
        pen.dot()

        # distance for another dot
        pen.forward(space)
    pen.backward(space*x)

    # direction
    pen.right(90)
    pen.forward(space)
    pen.left(90)




# Putting dots
pen.penup()
space = 85
print("Enter 3 for 3x3 Navagraha kolam")
dots = int(input())
draw(dots, space)


pen.hideturtle()


#open text file of kolam sequence


tkinter.Tk().withdraw() # prevents an empty tkinter window from appearing

file_path = filedialog.askopenfilename() #browse and select kolam sequence file

name = file_path #Store file path in a variable
file  = open(name) # open file using stored file path
sequence = file.read() #read the contents of  the text file


list = [] # Create an empty list
for l in range(0, len(sequence)):
    value = (sequence[l])
    list.append(value)
    l = l + 1


# Kolam sequence example

#list = ['b', 'a', 'c', 'c', 'c', 'd', 'd', 'c', 'a', 'b', 'b', 'a', 'a', 'a', 'c', 'd', 'd', 'c', 'c', 'd', 'b', 'a', 'a', 'b', 'b', 'b', 'd', 'd', 'd', 'c', 'c', 'd', 'b', 'a', 'a', 'b', 'z', 'z']


# Adding 'd' or 'b' at the beginning of the sequence if a kolam has its beginning at 90 degrees before the standard representation

if list[-3] == 'd':
    list.insert(0, 'd')
    pen.goto(-150, 200)
    pen.setheading(0)
    pen.penup()
    pen.forward(space/2)
else:
    list[-3] == 'b'
    list.insert(0, 'b')
    list.insert(-2, 'b')
    pen.goto(-150, 200)
    pen.setheading(-90)
    pen.penup()
    pen.forward(space/2)

# Condition for drawing circle shape

i = 0
while i < ((len(list)-3)):
    if list[i+3] != 'z' and list[i] != list[i+1] and list[i+1] != list[i+2] and list[i+2] != list[i+3]:

        c_element = list[i]
        f_element = list[i+1]
        s_element = list[i+2]
        t_element = list[i+3]
        if c_element == 'a' and f_element == 'b' and s_element == 'd' and t_element == 'c' :
            draw_circle_abdc(pen)
        elif c_element == 'b' and f_element == 'd' and s_element == 'c' and t_element == 'a':
            draw_circle_bdca(pen)
        elif c_element == 'd' and f_element == 'c' and s_element == 'a' and t_element == 'b':
            draw_circle_dcab(pen)
        elif c_element == 'c' and f_element == 'a' and s_element == 'b' and t_element == 'd':
            draw_circle_cabd(pen)
        elif c_element == 'b' and f_element == 'a' and s_element == 'c' and t_element == 'd':
            draw_circle_bacd(pen)
        elif c_element == 'a' and f_element == 'c' and s_element == 'd' and t_element == 'b':
            draw_circle_acdb(pen)
        elif c_element == 'c' and f_element == 'd' and s_element == 'b' and t_element == 'a':
            draw_circle_cdba(pen)
        else:
            c_element == 'd' and f_element == 'b' and s_element == 'a' and t_element == 'c'
            draw_circle_dbac(pen)
        i = i + 3


# condition for drawing U-shape

    elif list[i+2] != 'z' and list[i] != list[i+1] and list[i+1] != list[i+2]:
        c_element = list[i]
        f_element = list[i+1]
        s_element = list[i+2]
        if c_element == 'b' and f_element == 'd' and s_element == 'c':
            draw_U_bdc(pen)
        elif c_element == 'd' and f_element == 'c' and s_element == 'a':
            draw_U_dca(pen)
        elif c_element == 'c' and f_element == 'a' and s_element == 'b':
            draw_U_cab(pen)
        elif c_element == 'a' and f_element == 'b' and s_element == 'd':
            draw_U_abd(pen)
        elif c_element == 'b' and f_element == 'a' and s_element == 'c':
            draw_U_bac(pen)
        elif c_element == 'a' and f_element == 'c' and s_element == 'd':
            draw_U_acd(pen)
        elif c_element == 'c' and f_element == 'd' and s_element == 'b':
            draw_U_cdb(pen)
        else:
            c_element == 'd' and f_element == 'b' and s_element == 'a'
            draw_U_dba(pen)
        i = i + 2


# condition for drawing semicircle shape

    elif list[i] != list[i+1]:

        c_element = list[i]
        f_element = list[i+1]
        print(c_element)
        print(f_element)
        if c_element == 'b' and f_element == 'd':
            draw_semicircle_bd(pen)
        elif c_element == 'd' and f_element == 'c':
            draw_semicircle_dc(pen)
        elif c_element == 'c' and f_element == 'a':
            draw_semicircle_ca(pen)
        elif c_element == 'a' and f_element == 'b':
            draw_semicircle_ab(pen)
        elif c_element == 'a' and f_element == 'c':
            draw_semicircle_ac(pen)
        elif c_element == 'c' and f_element == 'd':
            draw_semicircle_cd(pen)
        elif c_element == 'd' and f_element == 'b':
            draw_semicircle_db(pen)
        else:
            c_element == 'b' and f_element == 'a'
            draw_semicircle_ba(pen)

        i = i + 1


# condition for drawing line shape

    elif list[i] == list[i + 1]:
        j = 0
        while j < (len(list)):
            if list[i+j] == list[i+j+1]:
                j = j + 1
            else:
                break


        k = 1
        c_element = list[i]
        while k < j:

            if c_element == 'a':
                draw_line_a(pen)
                k = k + 1
            elif c_element == 'b':
                draw_line_b(pen)
                k = k + 1
            elif c_element == 'c':
                draw_line_c(pen)
                k = k + 1
            else:
                c_element == 'd'
                draw_line_d(pen)
                k = k + 1

        i = i + j


# saving kolam image
# Uncomment the following to save the kolam design

#fname = "K1.eps"        # Change name to save a particular kolam design
#screen = Screen()
#screen.getcanvas().postscript(file=fname)

#print("Saved image to: ", fname)
#print("All done.")


# End
turtle.mainloop()

```
